```python
import pandas as pd
import matplotlib as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data/music_audio_data.csv', index_col=0)
```

```python
df.head(10)
```

```python
# Data Clean
```

```python
# remove filename & length since they are useless
df = df.drop(['filename', 'length'], axis=1)
```

```python
# detect null value and replace that with medium
for col in df.columns:
    missing = df[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0:
        print("Detect missing value in {}, fill missing with median".format(col))
        df[col].fillna(value=df[col].median())
    else:
        print("No missing value in {}".format(col))
```

```python
# find outliers
df["chroma_stft_mean"].describe()
```

```python
from scipy.stats.mstats import winsorize
```

```python
for column in df.columns:
    df[column] = winsorize(df[column], limits=[0.05, 0.05])
```
